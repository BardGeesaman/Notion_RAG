# Getting Started with Amprenta RAG

This notebook demonstrates how to use the `amprenta-client` library to interact with the RAG platform.

```python
from amprenta_rag.client import RAGClient

# Connect to API (auto-configured in JupyterHub)
import os
api_url = os.environ.get('API_URL', 'http://host.docker.internal:8000')
client = RAGClient(api_url=api_url)
print(f'Connected to {api_url}')
```

## List Experiments

```python
experiments = client.experiments.list()
print(f'Found {len(experiments)} experiments')
for exp in experiments[:5]:
    print(f'  - {exp.name[:60]}...')
```

## List Datasets

```python
datasets = client.datasets.list()
print(f'Found {len(datasets)} datasets')
for ds in datasets[:5]:
    print(f'  - {ds.name[:60]}...')
```

## Working with Compounds

```python
compounds = client.compounds.list(limit=5)
print(f'Found {len(compounds)} compounds')
for c in compounds:
    print(f'  - {c.name}: {c.smiles[:30] if c.smiles else "no SMILES"}...')
```
