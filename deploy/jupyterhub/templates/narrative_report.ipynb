# Narrative Report

Auto-generated narrative report using Amprenta evidence data.



```python
# Papermill parameters
entity_type = "program"
entity_id = ""
format = "html"

```

```python
from datetime import datetime
from uuid import UUID
from IPython.display import Markdown, display

from amprenta_rag.reporting.evidence_report import (
    generate_program_report,
    generate_dataset_report,
    generate_signature_report,
    generate_feature_evidence_report,
    format_evidence_report,
)

```

```python
def load_report(entity_type: str, entity_id: str):
    entity_type_lower = entity_type.lower() if entity_type else ""
    if entity_type_lower == "program":
        return generate_program_report(UUID(entity_id))
    if entity_type_lower == "dataset":
        return generate_dataset_report(UUID(entity_id))
    if entity_type_lower == "signature":
        return generate_signature_report(UUID(entity_id))
    if entity_type_lower == "feature":
        # Default feature_type to gene; adjust upstream as needed
        return generate_feature_evidence_report(
            feature_name=entity_id, feature_type="gene"
        )
    raise ValueError(f"Unsupported entity_type: {entity_type}")


report = load_report(entity_type, entity_id)
md_report = format_evidence_report(report)

```

```python
summary_md = """## Executive Summary

"""
summary_md += f"- Entity type: {entity_type}\n"
summary_md += f"- Entity ID: {entity_id}\n"
summary_md += f"- Generated at: {datetime.utcnow().isoformat()}Z\n\n"
summary_md += "## Data Provenance\n\n"
summary_md += "Data sourced from Postgres. Evidence assembled via cross-omics summaries.\n\n"
summary_md += "## Key Findings\n\nSee Evidence Report section below.\n\n"
summary_md += "## Visualizations\n\n(Reserved for future plots and charts.)\n\n"
summary_md += "## Methods\n\n- Reports generated via evidence_report utilities.\n- Validation/metrics pulled where available.\n\n"
summary_md += "## Next Steps\n\n- Review matched datasets and signatures.\n- Explore detailed evidence in the platform.\n"


display(Markdown(summary_md))
display(Markdown("## Evidence Report\n" + md_report))

```
